```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./results/score.csv')
```

```python
def get_rows_count_above_score(df, score1, score2):
    above_avg_scores = df.loc[df.loc[:, 'scores'].between(score1, score2), : ]
    return (len(above_avg_scores) / len(df)) * 100
```

```python
x = [i for i in range(1, 11)]
counts = [get_rows_count_above_score(df, (i) / 10, (i+1) / 10) for i in range(0, 10)]
```

```python
plt.style.use('seaborn')

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.75,0.75]) 

ax.bar(x, counts, linewidth=2.0)

plt.xlabel("Similarity Score (%)")
plt.ylabel("No. of Articles (%)")
plt.xticks(x, [f'{i*10} - {(i+1)*10}' for i in range(0,10)])

fig.savefig("./results/scores.png")

plt.show()
```
